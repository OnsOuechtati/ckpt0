```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
new_data= pd.read_csv('titanic-passengers.csv', sep=';')
new_data
```

```python
dics={"Yes":1 , "No":0} 
new_data['Survived']=new_data['Survived'].replace(dics)   #convert Survived variable into numerical
dicss={"male":1 , "female": 0}
new_data['Sex']=new_data['Sex'].replace(dicss)
new_data
```

```python
new_data['Age']=new_data['Age'].fillna(new_data['Age'].median())
new_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

#features extraction
x = new_data[['Age', 'Sex']]
y = new_data['Survived']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)  #splitting data with test size of 25%

logreg = LogisticRegression()   #build our logistic model
logreg.fit(x_train, y_train)  #fitting training data
y_pred  = logreg.predict(x_test)    #testing model’s performance
print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
```

```python
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
