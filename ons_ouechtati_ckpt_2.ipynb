```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns


```

```python
data1= pd.read_csv('titanic-passengers.csv', sep=';')
```

```python
data1
```

```python
pd.isnull(data1).sum()
```

```python
data1['Age']=data1['Age'].fillna(data1['Age'].median())
data1
```

```python
data1['Embarked'].value_counts()
```

```python
data1['Embarked']=data1['Embarked'].fillna(data1['Embarked'].mode()[0])
data1
```

```python
data1.drop('Cabin', axis=1, inplace=True)
```

```python
pd.isnull(data1).sum()
```

```python
data1.info()
```

```python
def plot_correlation_map( df ):
  plot_correlation_map( data1 )
```

```python
dics={"Yes":1 , "No":0} 
data1['Survived']=data1['Survived'].replace(dics)
dicss={"male":1 , "female": 0}
data1['Sex']=data1['Sex'].replace(dicss)
data1
```

```python
data1.groupby(['Pclass'])['Survived'].mean()
```

```python
data1['title']=data1['Name'].apply(lambda x:x.split(',')[1].split('.')[0].strip())
data1.drop('Name', axis=1, inplace=True)
data1
```

```python
plt.title("histogramme of different ages")
plt.xlabel("Age")
data1['Age'].plot.hist()
```

```python
plt.xlabel("Age")
plt.ylabel("Age value count")
plt.title("Bar plot of Ages")
vc=data1['Age'].value_counts()
vc.plot.bar(rot=45)
```
